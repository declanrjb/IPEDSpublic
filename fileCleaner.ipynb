```python
import os
import pandas as pd
```

```python
files = pd.DataFrame({
    'name':os.listdir('raw_data/2023')
})
```

```python
files['extension'] = files['name'].apply(lambda x: x.split('.')[1])
files['stem'] = files['name'].apply(lambda x: x.split('.')[0])
```

```python
list(files['name'])
```

```python
files['type'] = files['stem'].apply(lambda x: x.split('_')[-1])
```
